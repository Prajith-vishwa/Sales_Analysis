```python
# Load cleaned dataset
df = pd.read_csv("Luxury_Housing_Cleaned.csv")

# PostgreSQL connection details
username = "postgres"
password = "Praji"
host = "localhost"
port = "5432"
database = "luxury_housing_db"

# Create the database connection URL
# The format is: postgresql://username:password@host:port/database_name
db_url = f"postgresql://{username}:{password}@{host}:{port}/{database}"

# Define the file path for the clean data
csv_file_path = 'clean_data.csv'
table_name = 'property_bookings'

try:
    # 1. Create a database engine
    print("Connecting to the database...")
    engine = create_engine(db_url)
    print("Connection successful.")

    # 2. Read the clean data from CSV into a pandas DataFrame
    print(f"Reading data from {csv_file_path}...")
    df = pd.read_csv(csv_file_path)
    print("Data loaded into DataFrame:")
    print(df.head())

    # 3. Insert data into the PostgreSQL table
    # - `name`: The name of the SQL table.
    # - `con`: The database connection (engine).
    # - `if_exists='append'`: Add new data. Don't drop the table.
    # - `index=False`: Do not write the DataFrame index as a column.
    print(f"Inserting data into '{table_name}' table...")
    df.to_sql(table_name, con=engine, if_exists='append', index=False)

    print("\nData insertion complete!")

except Exception as e:
    print(f"An error occurred: {e}")

finally:
    if 'engine' in locals() and engine is not None:
        engine.dispose()
        print("Database connection closed.")
```

```python
# Load dataset
df = pd.read_csv(r"C:\Users\rpraj\OneDrive\Desktop\Luxury Housing Sales\Luxury_Housing_Cleaned.csv")

# Take first 20 rows
sample_df = df.head(20)

# Generate INSERT statements
queries = []
for _, row in sample_df.iterrows():
    query = f"""
    INSERT INTO luxury_housing (
        micro_market, project_name, developer_name, unit_size_sqft, configuration,
        ticket_price_cr, transaction_type, buyer_type, purchase_quarter,
        connectivity_score, amenity_score, possession_status, sales_channel,
        nri_buyer, nri_buyer_flag, locality_infra_score, avg_traffic_time_min,
        buyer_comments, price_per_sqft
    ) VALUES (
        '{row['Micro_Market']}', '{row['Project_Name']}', '{row['Developer_Name']}',
        {row['Unit_Size_Sqft'] if pd.notnull(row['Unit_Size_Sqft']) else 'NULL'},
        '{row['Configuration']}',
        {row['Ticket_Price_Cr'] if pd.notnull(row['Ticket_Price_Cr']) else 'NULL'},
        '{row['Transaction_Type']}', '{row['Buyer_Type']}', '{row['Purchase_Quarter']}',
        {row['Connectivity_Score'] if pd.notnull(row['Connectivity_Score']) else 'NULL'},
        {row['Amenity_Score'] if pd.notnull(row['Amenity_Score']) else 'NULL'},
        '{row['Possession_Status']}', '{row['Sales_Channel']}', '{row['NRI_Buyer']}',
        {row['NRI_Buyer_Flag'] if pd.notnull(row['NRI_Buyer_Flag']) else 'NULL'},
        {row['Locality_Infra_Score'] if pd.notnull(row['Locality_Infra_Score']) else 'NULL'},
        {row['Avg_Traffic_Time_Min'] if pd.notnull(row['Avg_Traffic_Time_Min']) else 'NULL'},
        '{str(row['Buyer_Comments']).replace("'", "''")}',
        {row['Price_per_Sqft'] if pd.notnull(row['Price_per_Sqft']) else 'NULL'}
    );
    """
    queries.append(query)

# Save queries to file (optional)
with open("insert_queries.sql", "w", encoding="utf-8") as f:
    f.writelines(queries)

# Print first 3 queries as sample
for q in queries[:10]:
    print(q)
```
